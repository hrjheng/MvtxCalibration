```python
import numpy as np

# read a text file and return a list of lines
def read_file(filename):
# Define an empty dictionary to store the data
    data = {}

    # Open the file for reading
    with open(filename, 'r') as file:
        # Initialize variables to store the current FELIX number and packet details
        current_felix = None
        current_packet = None
        current_feeids = None

        # Read each line in the file
        for line in file:
            # Strip any leading/trailing whitespace
            line = line.strip()
            
            # Check if the line starts with 'FELIX'
            if line.startswith('FLX'):
                # Extract the FELIX number
                current_felix = line.split()[0]
                # Initialize an empty dictionary for the current FELIX
                data[current_felix] = {}
            
            # Check if the line starts with 'Packet ids'
            elif line.startswith('Packet ids'):
                # Extract the packet ids
                packet_ids = [int(x) for x in line.split(':')[1].split()]
                # Assign packet ids to the current FELIX
                data[current_felix]['Packet_ids'] = packet_ids
                # Initialize an empty dictionary for packet details
                data[current_felix]['Packets'] = {}
            
            # Check if the line starts with 'Packet'
            elif line.startswith('Packet'):
                # Extract the packet number
                current_packet = int(line.split()[1])
                # Initialize an empty list for feeids
                current_feeids = []
            
            # Check if the line starts with 'Feeids'
            elif line.startswith('Feeids'):
                # Extract the feeids
                feeids = [int(x) for x in line.split(':')[1].split()]
                # Assign feeids to the current packet
                data[current_felix]['Packets'][current_packet] = feeids

    return data

filedat = read_file('FELIX_FEE_MAPS.txt')
# Print the dictionary
# print(filedat)
for felix_number in filedat.keys():
    # Print the information in the required format
    print("#", felix_number)
    print("if sys.argv[2] == \"" + felix_number + "\":")
    print("    PACKETS = ", filedat[felix_number]['Packet_ids'])

    print("    INPUTS = [")
    for packet_id, feeids in filedat[felix_number]['Packets'].items():
        print("        (", end="")
        print(*feeids, sep=", ", end="")
        print("),")
    print("    ]")

```
